```python
df = df.sort_values(by=['year', 'month'], ascending=True)
plt.figure(figsize=(6,2))
fig = px.pie(df, values='Price', names='year')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%Y")
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month_name()
```

```python
df_agg = df.groupby(['year', 'month']).agg({'Price': 'mean'}).reset_index()
df_pivot = df_agg.pivot(index='month', columns='year', values='Price')
plt.figure(figsize=(10, 6))
sns.heatmap(df_pivot, annot=True, fmt='.0f', linewidth=0.2, cmap='Blues')
plt.title('Price by Year and Month')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
